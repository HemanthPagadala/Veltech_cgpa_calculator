```python
import pytesseract
import cv2
# read the image file
img = cv2.imread("/content/drive/MyDrive/result55.jpg")
# extract the text from the image using OCR
text_output = pytesseract.image_to_string(img)
lines = text_output.splitlines()
#print(lines)
```

```python
IT=0;
words = []
grades = []
for line in lines:
    word = ""
    for i in range(len(line)):
        if line[i:i+3] == "115":
            word = "115"
            j = i + 3
            while j < len(line) and line[j].isalnum() and len(word) < 9:
                word += line[j]
                j += 1
            words.append(word)
        elif line[i:i+4].upper() == "PASS":
            if line[i-2].upper() == "S":
                grades.append(10)
            elif line[i-2].upper() == "A":
                grades.append(9)
            elif line[i-2].upper() == "B":
                grades.append(8)
            elif line[i-2].upper() == "C":
                grades.append(7)
            elif line[i-2].upper() == "D":
                grades.append(6)

#print(f"Words starting with 115: {words}")
#print(f"Grades: {grades}")
```

```python
credit=[]
for word in words:
    c=course_to_credit(word)
    credit.append(c)
#print(credit)
```

```python
no_of_courses = len(credit)
no_of_credits = sum(credit)
total = 0
cgpa = 0
for each in range(no_of_courses):
    score = grades[each]
    each_credits = credit[each]
    total = total + score * each_credits
cgpa = total / no_of_credits
print('CGPA is:', cgpa)
```
